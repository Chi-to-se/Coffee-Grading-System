```python
from ultralytics import YOLO
import time
import sys

MODEL_PATH = "yolo26cls_best_model.pt"

model = YOLO(MODEL_PATH)

image_path = "/Users/chitipatpangma/Documents/Year_3/Data Science/Final/RES vs YOLO/A02_17.png"

# warmup once before timing
model(image_path, verbose=False)

start = time.perf_counter()
results = model(image_path, verbose=False)
elapsed = (time.perf_counter() - start) * 1000  # ms

for result in results:
    probs      = result.probs
    top1_class = result.names[probs.top1]
    top1_conf  = probs.top1conf.item()
    print(f"Class: {top1_class} | Confidence: {top1_conf:.4f} | Time: {elapsed:.2f} ms")
```
